```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.maximize_window()
driver.get("https://droom.in/")
time.sleep(2)
driver.find_element_by_xpath("/html/body/div[10]/section[2]/div/div[1]/figure[1]/a").click()
```

```python
href = []
city = ['Surat','Jaipur','Kolkata','Chennai','Gurgaon','pune','Hyderabad']
for i in city:
    loc = driver.find_element_by_xpath("//*[@id='root']/div/div[1]/div[1]/div[1]/div/div[2]/div/div[2]/div/div").click()
    input_ = driver.find_element_by_xpath("/html/body/div[4]/div[3]/div/div/div/div[2]/div/div[1]/div/div/input")
    input_.click()
    input_.clear()
    input_.send_keys(i)
    time.sleep(2)
    input_.send_keys(Keys.ENTER)
    time.sleep(3)
    
    for i in range(0,10):   
        actions = ActionChains(driver)
        actions.send_keys(Keys.PAGE_UP).perform()
        time.sleep(1)
    
    for i in range(0,8):
        actions = ActionChains(driver)
        for _ in range(0,5):
            actions.send_keys(Keys.PAGE_DOWN).perform()
            time.sleep(3)
        try:
            load_more = driver.find_element_by_xpath("//*[@id='root']/div/div[2]/div/div[3]/div/div[3]/button").click()
        except NoSuchElementException:
            pass

        link = driver.find_elements_by_xpath("//*[@id='root']/div/div[2]/div/div[3]/div/div[2]/div/div/div/div/div/h2/a")
        for i in link:
            href.append(i.get_attribute('href'))
    
```

```python
Name = []
Price = []
KM = []
Location = []
Owner = []
Year = []
fuel_type = []
car_type = []
Seat = []
gear_type = []
EMI = []
rating = []
review = []
mileage = []


for i in href[0:5000]:
    driver.get(i)
    time.sleep(2)
    actions = ActionChains(driver)
    for _ in range(0,1):
        actions.send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(2)
        
    for _ in range(0,1):
        actions.send_keys(Keys.PAGE_UP).perform()
        time.sleep(2)
        
    time.sleep(2)
    
    try:
        name = driver.find_element_by_xpath("//h1[@class='d-font-weight-500 d-font-size-22 d-text-dark d-margin-top-0']")
        Name.append(name.text)
    except NoSuchElementException:
        name2 = driver.find_element_by_xpath("//div[@class='d-font-size-22 d-font-weight-400 d-position-relative']")
        Name.append(name2.text)
    except NoSuchElementException:
        Name.append("-")
        
    try:
        price = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[2]/div[1]/div[1]/div[5]/div[1]/div[2]")
        Price.append(price.text)
    except NoSuchElementException:
        Price.append("-")
        
    try:
        km = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[1]/div")
        KM.append(km.text.split()[0])
    except NoSuchElementException:
        KM.append("-")
    try:
        location = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[2]/div")
        Location.append(location.text)
    except NoSuchElementException:
        Location.append("-")
    try:
        owner = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[3]/div")
        Owner.append(owner.text)
    except NoSuchElementException:
        Owner.append("-")
    try:
        year = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[5]/div")
        Year.append(year.text.split()[-1])
    except NoSuchElementException:
        Year.append("-")
    try:
        ft = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[6]/div")
        fuel_type.append(ft.text)
    except NoSuchElementException:
        fuel_type.append("-")
    try:
        ct = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[7]/div")
        car_type.append(ct.text)
    except NoSuchElementExcepton:
        car_type.append("-")
        
    try:
        seat = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[8]/div")
        Seat.append(seat.text.split()[0])
    except NoSuchElementException:
        Seat.append("-")
    try:
        gt = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[9]/div")
        gear_type.append(gt.text)
    except NoSuchElementException:
        gear_type.append("-")
        
    try:
        mil = driver.find_element_by_xpath("//ul[@class='d-display-table d-width-100']/li[10]/div")
        mileage.append(mil.text.split()[0])
    except NoSuchElementException:
        mileage.append("-")
        
    try:
        emi = driver.find_element_by_xpath("//div[@class='d-text-green d-font-size-13 d-margin-top-5 d-display-inline-block d-width-100 bestmatch-icon website']")
        EMI.append(emi.text.split()[-1])
    except NoSuchElementException:
        EMI.append("-")
    try:
        rat = driver.find_element_by_xpath("//*[@id='rating_score']")
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append("-")
    try:
        rev = driver.find_element_by_xpath("//*[@id='review_count']")
        review.append(rev.text)
    except NoSuchElementException:
        review.append("-")
    
```

```python
car_detail2 = pd.DataFrame({})
car_detail2['Name'] = Name
car_detail2['KM'] = KM
car_detail2['Location'] = Location
car_detail2['Owner'] = Owner
car_detail2['Year'] = Year
car_detail2['fuel_type'] = fuel_type
car_detail2['car_type'] = car_type
car_detail2['Seat'] = Seat
car_detail2['gear_type'] = gear_type
car_detail2['EMI'] = EMI
car_detail2['mileage'] = mileage
car_detail2['rating'] = rating
car_detail2['review'] = review
car_detail2['Price'] = Price
car_detail2
```

```python
car_detail2.to_csv('car_detail.csv')
```
